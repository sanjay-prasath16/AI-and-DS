```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing pre-built dataset in seaborn
crash_df = sns.load_dataset('car_crashes')
crash_df
```

```python
# distributed plots - helps to understand the underlying pattern of data
sns.displot(crash_df['not_distracted'])
```

```python
# join plot - shows the relationship of two variable or also show the relationship of each variable
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
# KDE Plots - Visualizes the probability density function (PDF) of a variable (1D or 2D), offering a smooth alternative to histograms.
sns.kdeplot(crash_df['alcohol'])
```

```python
# pair plots - relationships between the whole dataframe's numerical value
tips_df = sns.load_dataset('tips')

sns.pairplot(tips_df, hue='sex', palette='Blues')
```

```python
# Rug Plot - Adds small vertical/horizontal ticks at each data point, typically overlaid on a KDE or histogram. used to show individual observations
sns.rugplot(tips_df['tip'])
```

```python
# styling
sns.set_style('darkgrid')

plt.figure(figsize=(8, 4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
# Bar Plot
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)
```

```python
# count plot
sns.countplot(x='sex', data=tips_df)
```

```python
# Box Plot
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
# Violin Plot
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
# Strip Plot
plt.figure(figsize=(8, 5))

sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)
```

```python
# Sworm Plot
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

```python
# Palettes
plt.figure(figsize=(8,5))

sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='magma')
```

```python
# Matrix plots
# Heatmaps
plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)

# Select only numeric columns
crash_mx = crash_df.select_dtypes(include='number').corr()

sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
# Cluster Map - uncover patterns, similarities, or groupings in your data — both across rows and columns.
iris = sns.load_dataset('iris')
species = iris.pop('species')

sns.clustermap(iris)
```

```python
# Pair Grid
iris = sns.load_dataset('iris')
# iris_g = sns.PairGrid(iris, hue="species")
# iris_g.map_diag(plt.hist)
# iris_g.map_offdiag(plt.scatter)
iris_g = sns.PairGrid(iris, hue="species",
                     x_vars=["sepal_length", "sepal_width"],
                     y_vars=["petal_length", "petal_width"])
iris_g.map(plt.scatter)
iris_g.add_legend()
```

```python
# Facet Grid
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8)
```
