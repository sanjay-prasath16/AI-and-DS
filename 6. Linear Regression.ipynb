```python
# implementing multiple linear regression (multiple variables)
mf = pd.read_csv('./Data files/linearRegression/multiple_variable.csv')
mf.head()
```

```python
# pre processing the datas
import math
median_bathrooms = math.floor(mf.bedrooms.median())
median_bathrooms
```

```python
mf.bedrooms = mf.bedrooms.fillna(median_bathrooms)
mf
```

```python
# training the model
mulReg = linear_model.LinearRegression()
mulReg.fit(mf[['area', 'bedrooms', 'age']], mf.price)
```

```python
mulReg.coef_
```

```python
mulReg.intercept_
```

```python
mulReg.predict([[3000,3,40]])
```
