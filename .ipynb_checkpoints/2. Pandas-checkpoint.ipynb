```python
bios_new = bios.copy()
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
bios_new['first_name']
```

```python
bios_new.to_csv('./Data files/Pandas/bios_new.csv', index=False)
```

```python
# Merging and Concatenating
nocs = pd.read_csv('./Data files/Pandas/noc_regions.csv')
bios_new = pd.merge(bios, nocs, left_on='born_country', right_on='NOC', how='left')
bios_new.head()
```

```python
bios_new.rename(columns={'region': 'born_country_full'}, inplace=True)
bios_new.head()
```

```python
usa = bios[bios['born_country'] == 'USA'].copy()
fra = bios[bios['born_country'] == 'FRA'].copy()
new_df = pd.concat([usa,fra])
print(new_df.head())
```

```python
import numpy as np
coffee.loc[[0, 1], 'Units Sold'] = np.NAN
coffee
```

```python
coffee.fillna(coffee['Units Sold'].mean())
```

```python
# if NAN is inbetween like not at starting means we can use interpolate function to fill it with the numbers
```

```python
coffee[coffee['Units Sold'].isna()]
```

```python
coffee[coffee['Units Sold'].notna()]
```

```python
# Aggregate Functions
# value_counts() - used to get the count of the specifying value
coffee.groupby(['Coffee Type']).agg({'Units Sold': 'sum', 'price': 'mean'})
```

```python
# pivot table
pivot = coffee.pivot(columns='Coffee Type', index='Day', values='price')
pivot
```

```python
coffee['yes_revenue'] = coffee['price'].shift(2)
coffee
```

```python
bios['heighest_rank'] = bios['height_cm'].rank(ascending=False)
bios.sort_values(['heighest_rank'])
```

```python
coffee['cumulative_price'] = coffee['price'].cumsum()
coffee
```

```python
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()
latte['ThreeDays'] = latte['Units Sold'].rolling(2).sum()
latte
```
